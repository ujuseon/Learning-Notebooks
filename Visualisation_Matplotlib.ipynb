#### Круговая диаграмма (Pie chart)

matplotlib.pyplot.pie
params:
- x: 1D массив
- explode: смещение каждого сегмента; массив соответствующий x
- labels: массив меток для каждого сегмента 
- colors: массив цветов для каждого сегмента
- startangle: угол; начало диаграммы против часовой стрелки от оси X.



```python
y = np.array([35, 25, 25, 15])
mylabels = ["Cars", "Trains", "Planes", "Ships"]
mycolors = ['#9FE2BF', "#6495ED", "#CCCCFF", "#FF7F50"]
myexplode = [0.01, 0.01, 0.01, 0.01]

plt.pie(y, labels=mylabels, colors=mycolors, explode=myexplode)
plt.legend(title="Four Types:", bbox_to_anchor=(1, 1))
plt.title("Types of transportation")
plt.show()
```

#### Линейная диаграмма (Bar chart, Столбчатая диаграмма)

Измерения задаются по height и width.

params:
- x: Столбцы располагаются в точке x с заданным выравниванием.
- height: высота столбцов. 
- width: ширина столбцов
доп. параметры:
- edgecolor: цвет границ столбцов

В Matplotlib **patches** это коллекция графических объектов, представляющих собой фигуры, такие как прямоугольники, круги, многоугольники и т. д. Когда вы создаете график с помощью Matplotlib, каждый элемент в графике, такой как столбцы в столбчатой диаграмме, линии в линейном графике или патчи в коллекции патчей, хранятся как отдельные объекты.

В контексте столбчатой диаграммы каждый столбец представляется как патч-прямоугольник. Атрибут **patches** оси (ax.patches) предоставляет доступ к этим отдельным патчам.

```python
data = {'Math': 30, 'Science': 25, 'English': 20, 'History': 22}
courses = list(data.keys())
values = list(data.values())

plt.bar(courses, values, color='pink', width=0.5)
plt.xlabel("Courses offered")
plt.ylabel("Number of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
data = {'car': ['Ferrari', 'Lamborghini', 'Porsche', 'Aston Martin', 'Bentley', 'Rolls-Royce', 'Maserati', 'Jaguar', 'AMG', 'Audi', 'BMW', 'Mercedes'],
        'price': [350, 320, 280, 250, 230, 210, 190, 180, 170, 160, 150, 140]}
df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize =(7, 5))

ax.barh(df['car'], df['price'], color='#660099')

# Удаляем границы графика
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Удаление x, y Ticks
# "Тики" (ticks) - это метки на осях графика, которые обозначают конкретные значения данных. 
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Добавление отступа между осями и метками
ax.xaxis.set_tick_params(pad = 5) # price
ax.yaxis.set_tick_params(pad = 10) # car

# Добавление к графику сетки с пунктирным стилем, шириной линии 0,5 и альфа (прозрачностью) 0,2.
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
ax.grid(True, linestyle='dotted', linewidth=0.5, alpha=0.2)

# Ось Y инвертируется, так что самые высокие значения находятся в верхней части графика
ax.invert_yaxis()

# Добавление подписей к столбцам
# str(round((i.get_width()), 2)) преобразует ширину полосы в строку и округляет ее до двух знаков. Это будет текст аннотации.
for i in ax.patches:
    plt.text(i.get_width()+5.2, i.get_y()+0.5, 
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

ax.set_title('Sports car and their prices in $k')

plt.show()
```

#### Гистограмма (Histogram) 



```python
import matplotlib.pyplot as plt
import numpy as np

barWidth = 0.25
fig, ax = plt.subplots(figsize=(7, 5))

fiction = [10, 40, 150, 60, 75]
non_fiction = [90, 61, 70, 121, 30]
fantasy = [80, 100, 40, 75, 155]

br1 = np.arange(len(fiction))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
    
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

plt.bar(br1, fiction, color='#FF9999', width=barWidth, edgecolor='grey', label='Fiction')
plt.bar(br2, non_fiction, color='#99CC99', width=barWidth, edgecolor='grey', label='Non-Fiction')
plt.bar(br3, fantasy, color='#99CCFF', width=barWidth, edgecolor='grey', label='Fantasy')

ax.grid(True, linestyle='dotted', linewidth=0.5, alpha=0.6)

plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Books Borrowed', fontsize=12)
plt.xticks([r + barWidth for r in range(len(fiction))],
           ['2019', '2020', '2021', '2022', '2023'])
plt.legend()
plt.title('Number of Books Borrowed by Genre Over the Years')
plt.show()
```

#### Линейный график (Line plot)

Линеный график обычно используется для визуализации временных рядов.

matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, \**kwargs)

Параметр *args используется для передачи переменного числа позиционных аргументов. Это означает, что можно передавать любое количество аргументов при вызове функции plot(), и они будут интерпретироваться как данные для построения графика.

Относительное расположение *args и \**kwargs в сигнатуре функции важно, поскольку *args собирает все позиционные аргументы в кортеж, а \**kwargs собирает все именованные аргументы в словарь. Поэтому *args обычно размещается в начале сигнатуры, чтобы сначала собрать все позиционные аргументы, а затем оставить оставшиеся аргументы для \**kwargs.

Простой линейный график

```python
x = np.linspace(1, 100, 10)
y = np.log(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('logx')
plt.title('Простой линейный график')

# Устанавливаем пределы осей x и y на графике
plt.xlim(0, 100)
plt.ylim(0, 5)
plt.show()
```

Несколько графиков на одном рисунке

```python
x = np.linspace(1, 20, 200)

plt.plot(x, np.cos(x))
plt.plot(x, np.log(x))

plt.xlabel('x')
plt.title('Sin and Log')
plt.xlim(1,10)
plt.ylim(-1.0 , 2.5)
plt.show()
```

Построение графиков по разным осям.

Функция **twinx()** в библиотеке Matplotlib создает вторую y-ось на графике, которая является параллельной основной y-оси. Это позволяет отображать два набора данных с разными диапазонами значений по оси y на одном графике, что особенно полезно, когда значения данных имеют значительное различие в шкале.
